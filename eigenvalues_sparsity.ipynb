# For a = 1

```python
import numpy as np 
import matplotlib.pyplot as plt

n = 200;
a = 1;
tol = 1e-10
A = np.random.randn(n,n)
A[np.where(A <a)] = a
A = A - np.full((n,n),a)
A = A - np.diag (np.diag (A));
A_colsum = np.sum(A, axis=0)
A_colsum[np.where(A_colsum<tol)] = tol
L=np.dot(A, np.diag(np.reciprocal (A_colsum)))
```

## Showing S = kL + (1- k)E is column stochastic


```python
E = np.full((200,200),1/n)
k = 0.85
S = k*L + (1 - k)*E

# Checking if S is column stochastic

print(np.sum(S,axis=0))
```

```python
#Computing Eigenvalues
e = np.linalg.eigvals(S)

fig, ax = plt.subplots()
plt.scatter(e.real, e.imag, s=16)
unit_circle = plt.Circle((0.0, 0.0), 1,fill=False)

plt.xlim(-2,2)
plt.ylim(-2,2)
ax.add_patch(unit_circle)
plt.title('a= '+str(a)+', k = '+str(k));
```

```python
#Looking at various k's

k_test = [0.1, 0.5, 0.99]

for i in k_test:
  S = i*L + (1 - i)*E
  e = np.linalg.eigvals(S)

  fig, ax = plt.subplots()
  plt.scatter(e.real, e.imag, s=16)
  unit_circle = plt.Circle((0.0, 0.0), 1,fill=False)

  plt.xlim(-2,2)
  plt.ylim(-2,2)
  ax.add_patch(unit_circle)
  plt.title('a= '+str(a)+', k = '+str(i));
```
